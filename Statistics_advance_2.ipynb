**1. What is hypothesis testing in statistics?**
- Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps you determine whether a specific claim (or hypothesis) about a population parameter is supported by the data or not.
- Common Types of Tests
	-	z-test: When population variance is known and sample size is large
	-	t-test: When population variance is unknown and sample size is small
	-	Chi-square test: For categorical data (like testing independence)
	-	ANOVA: For comparing more than two group means


**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- Hypothesis testing involves two competing statements: the null hypothesis (H₀) and the alternative hypothesis (H₁ or Hₐ). These are used to make decisions about a population based on sample data.
- Null Hypothesis (H₀)
	-	It is the default or original assumption.
	-	It states that there is no effect, no difference, or no change.
	-	It is considered true unless strong evidence suggests otherwise.
	-	It is the statement being tested in a hypothesis test.
- Key Differences
	-	H₀ assumes no change; H₁ suggests a change or effect.
	-	H₀ includes equality signs ( =, ≤, ≥ ); H₁ uses inequality signs ( ≠, <, > ).
	-	H₀ is presumed true unless evidence leads to its rejection; H₁ is considered when H₀ is rejected.
	-	Hypothesis testing focuses on finding enough evidence to reject H₀, not directly proving H₁.

**3. What is the significance level in hypothesis testing, and why is it important?**
-	The significance level (denoted as α) is the threshold used to decide whether to reject the null hypothesis in a statistical test.
-	It represents the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.
-	Common values for α are 0.05, 0.01, and 0.10. For example, if α = 0.05, there is a 5% risk of wrongly rejecting a true null hypothesis.
-	The significance level is chosen before conducting the test to control the likelihood of making false discoveries.
	•	After performing the test, the p-value is compared with α:
	-	If p-value < α, reject the null hypothesis (evidence is statistically significant).
	-	If p-value ≥ α, do not reject the null hypothesis (evidence is not strong enough).
-	The smaller the α, the stronger the evidence must be to reject H₀.
-	Choosing an appropriate α is important because it affects the reliability of your conclusions.
-	In sensitive fields like medicine or public safety, a smaller α (like 0.01) is often preferred to reduce the chance of false positives.


**4. What does a P-value represent in hypothesis testing?**
	-	It helps measure the strength of evidence against the null hypothesis.
	-	A small p-value indicates that the observed data is unlikely under H₀, suggesting that H₀ may not be true.
	-	A large p-value means the observed result is likely under H₀, so there’s not enough evidence to reject it.
	-	The p-value is compared with the significance level (α) to make a decision:
	-	If p-value < α, reject H₀ (evidence is statistically significant).
	-	If p-value ≥ α, do not reject H₀ (evidence is not strong enough).
	-	The p-value does not tell you the probability that H₀ is true or false — it only tells you how compatible the data is with H₀.
	-	For example, a p-value of 0.03 means there is a 3% chance of observing such data (or more extreme) if H₀ is true.

**5. How do you interpret the P-value in hypothesis testing?**
- The p-value tells you how likely it is to get the observed result, or something more extreme, if the null hypothesis (H₀) is true.
-	It is a measure of evidence against H₀.
-	A small p-value means the observed result is unlikely under H₀ and gives stronger evidence against H₀.
-	A large p-value means the observed result is likely under H₀ and gives weaker evidence against H₀.
-	The interpretation depends on the chosen significance level (α), such as 0.05 or 0.01:
 -	If p-value < α → reject H₀ → the result is statistically significant.
 -	If p-value ≥ α → do not reject H₀ → the result is not statistically significant.
-	Example interpretation:
 -	If p-value = 0.02 and α = 0.05, you reject H₀ and conclude that there is strong evidence supporting the alternative hypothesis (H₁).
 -	If p-value = 0.25 and α = 0.05, you do not reject H₀ because the evidence is not strong enough.
-	Keep in mind, a p-value does not tell you the probability that H₀ is true — it only tells you how consistent the data is with H₀.

**6. What are Type 1 and Type 2 errors in hypothesis testing?**
-	Type 1 Error (False Positive)
	-	This occurs when you reject the null hypothesis (H₀) even though it is actually true.
	-	You are seeing an effect that doesn’t really exist.
	-	The probability of making a Type 1 error is equal to the significance level α (e.g., 0.05 means a 5% chance).
	-	Example: A medical test shows a disease is present when the person is actually healthy.
	-	Type 2 Error (False Negative)
	-	This occurs when you fail to reject the null hypothesis even though the alternative hypothesis is true.
	-	You are missing a real effect that actually exists.
	-	The probability of making a Type 2 error is denoted by β.
	-	The power of a test is 1 − β, which represents the probability of correctly rejecting H₀.
	-	Example: A medical test fails to detect a disease when the person actually has it.

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- 	A one-tailed test checks for an effect in only one direction
- either greater than or less than a certain value, but not both.
-	A two-tailed test checks for an effect in both directions.
- it tests whether the parameter is different from a certain value, either higher or lower

**8. What is the Z-test, and when is it used in hypothesis testing?**
	-	The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means) when the population standard deviation is known.
	-	It is based on the standard normal distribution (Z-distribution).
  -	When to use the Z-test
	 -	The sample size is large (typically n ≥ 30).
	 -	The population standard deviation (σ) is known.
	 - The data is approximately normally distributed, or the sample is large enough for the Central Limit Theorem to apply.
-	Types of Z-tests
	-	One-sample Z-test: compares the sample mean to a known population mean.
	-	Two-sample Z-test: compares the means of two independent samples.
	-	Z-test for proportions: compares sample proportions to population proportions.

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**
-	The Z-score (also called the standard score) tells you how many standard deviations a data point or sample mean is from the population mean under the null hypothesis.
-	In hypothesis testing, it helps determine whether to reject the null hypothesis by comparing it to critical values from the standard normal distribution.

**10. What is the T-distribution, and when should it be used instead of the normal distribution?**

-	The T-distribution (or Student’s t-distribution) is a probability distribution used in statistics for estimating population parameters when the sample size is small and the population standard deviation is unknown.
-	It looks similar to the normal distribution but has heavier tails, which means it accounts for more variability, especially with smaller samples.

**11. What is the difference between a Z-test and a T-test?**
-	The Z-test and T-test are both used to test hypotheses about population means, but they differ based on what information is available and the size of the sample.
-	Z-test
	-	Used when the population standard deviation (σ) is known.
	-	Typically used when the sample size is large (n ≥ 30).
	-	Assumes the data is normally distributed or sample size is large enough for the Central Limit Theorem to apply.
	-	Based on the standard normal distribution.
	-	Example: Testing if a sample mean differs from a known population mean when σ is known.
-	T-test
	-	Used when the population standard deviation is unknown.
	-	Typically used when the sample size is small (n < 30).
	-	Also assumes the data is approximately normal.
	-	Based on the T-distribution, which has heavier tails than the normal distribution.
	-	Example: Testing if a sample mean differs from a population mean when you only have the sample standard deviation.

**12. What is the T-test, and how is it used in hypothesis testing?**
 -	The T-test is a statistical test used to compare the mean(s) of one or two samples when the population standard deviation is unknown.
 -	It is commonly used when dealing with small sample sizes (n < 30) and when the data is approximately normally distributed.
 -	The T-test is based on the T-distribution, which adjusts for extra variability in small samples.

**13. What is the relationship between Z-test and T-test in hypothesis testing?**
-	The Z-test and T-test are both used to test hypotheses about population means, but they are chosen based on different conditions.
-	Both tests follow a similar logic: compare the sample mean to a hypothesized population mean, then determine if the difference is statistically significant.
-	Similarities
	-	Both are used in hypothesis testing for means.
	-	Both involve calculating a test statistic (Z or t) to evaluate how far the sample mean is from the population mean.
	-	Both compare the test statistic to a critical value or use the p-value to make a decision.
	-	Both rely on the assumption that the data is approximately normal or the sample size is large enough.

**14. What is a confidence interval, and how is it used to interpret statistical results?**
-	A confidence interval (CI) is a range of values that is likely to contain the true value of a population parameter (like the mean or proportion) with a certain level of confidence.
-	It gives an estimate of the parameter along with an associated level of uncertainty.
-	Confidence level
	-	Common levels: 90%, 95%, 99%
	-	A 95% confidence level means that if we repeated the experiment many times, about 95% of the confidence intervals calculated would contain the true parameter
	-	Higher confidence → wider interval (more certainty but less precision)
	-	Lower confidence → narrower interval (less certainty but more precision)

**15. What is the margin of error, and how does it affect the confidence interval?**
-	The margin of error (ME) is the amount added and subtracted from a sample estimate to create a confidence interval.
-	It reflects the maximum expected difference between the sample statistic and the true population parameter due to random sampling.
	-	Formula for margin of error (for means)
	-	ME = critical value × (standard deviation / √n)
	-	The critical value depends on the confidence level (e.g., 1.96 for 95% with a Z-distribution)
	-	A larger standard deviation or a smaller sample size will increase the margin of error
	-	How it affects the confidence interval
	-	Confidence interval = sample estimate ± margin of error
	-	A larger margin of error results in a wider confidence interval, showing less precision but more certainty
	-	A smaller margin of error gives a narrower confidence interval, showing more precision but less certainty

**16. How is Bayes' Theorem used in statistics, and what is its significance**
-	Bayes’ Theorem is a mathematical formula used to update the probability of a hypothesis based on new evidence.
-	It is a key concept in Bayesian statistics, where probabilities represent degrees of belief or uncertainty rather than fixed frequencies.
-	Bayes’ Theorem formula
	-	P(A | B) = [P(B | A) × P(A)] / P(B)
	-	P(A | B) = probability of event A given B (posterior probability)
	-	P(B | A) = probability of event B given A (likelihood)
	-	P(A) = initial probability of A (prior)
	-	P(B) = total probability of event B (normalizing constant)
-	How it’s used in statistics
	-	To revise predictions or beliefs in light of new data
	-	Common in fields like medical diagnosis, machine learning, spam filtering, and risk assessment
	-	Helps in making probabilistic inferences even with incomplete or uncertain data
-	Example
	-	Suppose a disease affects 1 in 1000 people (P(Disease) = 0.001)
	-	A test has 99% sensitivity and 95% specificity
	-	A person tests positive — Bayes’ Theorem helps calculate the actual probability they have the disease, factoring in false positives
	-	This is more informative than just knowing the test is “99% accurate”
	-	Significance of Bayes’ Theorem
	-	It shifts focus from “what is the chance of this data if the hypothesis is true?” to “what is the chance the hypothesis is true given this data?”
	-	Enables continuous learning: beliefs are updated as more evidence is gathered
	-	Core to Bayesian inference, where parameters are treated as random variables with their own distributions

**17. What is the Chi-square distribution, and when is it used?**
-	The Chi-square (χ²) distribution is a probability distribution used in statistics to test hypotheses about categorical data.
-	It is asymmetric, only takes non-negative values, and depends on a parameter called degrees of freedom (df).
-	When is the Chi-square distribution used?
	-	When analyzing frequency counts in categories
	-	When testing the independence between two categorical variables
	-	When testing the goodness-of-fit between observed and expected distributions
	-	When evaluating the variance of a normal population (in some cases)

-	When to use this test
	-	You have one categorical variable with two or more categories
	-	You want to test whether the observed distribution matches a hypothesized distribution
	-	The data consists of counts or frequencies, not percentages or means
-	Steps to apply the test
	-	Define your null and alternative hypotheses
	-	Collect the observed frequencies from your data
	-	Determine the expected frequencies under H₀ (usually based on equal distribution or a known proportion)
	-	Calculate the χ² statistic using the formula
	-	Find the critical value from the Chi-square distribution table based on degrees of freedom = (number of categories − 1)
	-	Compare the test statistic with the critical value or compute the p-value
	-	If χ² > critical value or p-value < α → reject H₀
	-	Otherwise → do not reject H₀

**19. What is the F-distribution, and when is it used in hypothesis testing?**
-	The F-distribution is a continuous probability distribution that arises when comparing variances of two populations or testing the overall significance in ANOVA (Analysis of Variance).
-	It is right-skewed and defined by two types of degrees of freedom: one for the numerator and one for the denominator.
-	When is the F-distribution used?
	-	When testing whether two population variances are equal
	-	When conducting ANOVA to compare the means of three or more groups
	-	In regression analysis to test the overall significance of a model

**20. What is an ANOVA test, and what are its assumptions?**
-	ANOVA (Analysis of Variance) is a statistical method used to test whether there are significant differences between the means of three or more independent groups.
-	Instead of comparing means pairwise (which increases error), ANOVA checks if at least one group mean is different from the others using a single F-test.
-	When to use ANOVA
	-	When comparing the means of three or more groups
	-	When the data is numerical and the groups are categorical and independent
	-	Example: Testing if students taught using three different methods have different average test scores

**21. What are the different types of ANOVA tests?**
-	One-way ANOVA
	-	Used to compare the means of three or more independent groups based on one independent variable (factor).
	-	Example: Comparing test scores of students taught by 3 different teachers.
	-	H₀: All group means are equal
	-	H₁: At least one group mean is different
-	Two-way ANOVA
	-	Used to examine the effect of two independent variables (factors) on a single dependent variable.
	-	It also allows for testing interaction effects between the two factors.
	-	Example: Analyzing how different teaching methods and student gender affect test scores.
	-	Tests three hypotheses:
	-	Effect of factor A
	-	Effect of factor B
	-	Interaction between A and B
-	Repeated Measures ANOVA
	-	Used when the same subjects are measured multiple times under different conditions or at different times.
	-	Helps control for individual variability.
	-	Example: Measuring blood pressure of patients before, during, and after treatment.
	-	Similar to paired t-tests but for more than two conditions
-	MANOVA (Multivariate ANOVA)
	-	An extension of ANOVA when there are multiple dependent variables.
	-	Tests whether group differences exist across a combination of dependent variables.
	-	Example: Comparing student groups based on both math and science scores simultaneously
-	ANCOVA (Analysis of Covariance)
	-	Combines ANOVA and regression
	-	Used to compare group means while controlling for the effect of a covariate (a continuous control variable).
	-	Example: Comparing test scores among schools while controlling for socioeconomic status

**22. What is the F-test, and how does it relate to hypothesis testing?**
-	The F-test is a statistical test used to compare two variances or to assess whether multiple group means are significantly different, especially in ANOVA (Analysis of Variance).
-	It is based on the F-distribution, which is skewed and only takes positive values.
-	Hypotheses in F-tests
	-	For variance comparison:
	-	H₀: σ₁² = σ₂² (variances are equal)
	-	H₁: σ₁² ≠ σ₂² (variances are different)
	-	For ANOVA:
	-	H₀: All group means are equal
	-	H₁: At least one group mean is different
-	How it relates to hypothesis testing
	-	You calculate the F-statistic and compare it to the critical F-value from the F-distribution table (based on degrees of freedom)
	-	Or you compare the p-value to your significance level (α)
  - If p-value < α → reject H₀
	- If p-value ≥ α → do not reject H₀

**1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results?**

```python
import numpy as np
from scipy import stats

# Step 1: Define the sample data
sample_data = [105, 110, 102, 108, 115, 107, 109, 111, 106, 112]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 2: Define known population parameters
population_mean = 100
population_std_dev = 15  # Known population standard deviation

# Step 3: Calculate the Z-score
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Step 4: Calculate the two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 5: Print the results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_statistic, 3))
print("P-value:", round(p_value, 4))

# Step 6: Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("Interpretation: There is a statistically significant difference between the sample mean and the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("Interpretation: There is no statistically significant difference between the sample mean and the population mean.")
```

**4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Sample data
sample = [105, 110, 102, 108, 115, 107, 109, 111, 106, 112]
sample_mean = np.mean(sample)
sample_size = len(sample)

# Step 2: Known population parameters
population_mean = 100
population_std = 15  # Known population standard deviation

# Step 3: Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 5: Print results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_score, 3))
print("P-value:", round(p_value, 4))

# Step 6: Hypothesis test
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)  # Two-tailed

if abs(z_score) > z_critical:
    print("Conclusion: Reject the null hypothesis.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")

# Step 7: Plotting the decision regions
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Standard Normal Distribution", color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')

# Plot the Z-score
plt.axvline(z_score, color='black', linestyle='--', linewidth=2, label=f'Z = {z_score:.2f}')

# Formatting
plt.title("Two-Tailed Z-Test Decision Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

**5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    """
    Parameters:
    - mu0: Population mean under null hypothesis (H0)
    - mu1: Population mean under alternative hypothesis (H1)
    - sigma: Population standard deviation (assumed known)
    - n: Sample size
    - alpha: Significance level (Type I error rate)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (right-tailed test)
    z_alpha = norm.ppf(1 - alpha)
    critical_value = mu0 + z_alpha * se

    # Type II error (beta): Probability of failing to reject H0 when H1 is true
    beta = norm.cdf((critical_value - mu1) / se)

    # x-axis range
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)

    # Distributions under H0 and H1
    y_h0 = norm.pdf(x, mu0, se)
    y_h1 = norm.pdf(x, mu1, se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label="H₀ Distribution", color="blue")
    plt.plot(x, y_h1, label="H₁ Distribution", color="green")

    # Shade Type I error region (alpha)
    plt.fill_between(x, y_h0, where=(x >= critical_value), color='red', alpha=0.4, label="Type I Error (α)")

    # Shade Type II error region (beta)
    plt.fill_between(x, y_h1, where=(x < critical_value), color='orange', alpha=0.4, label="Type II Error (β)")

    # Critical value line
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    # Labels and legend
    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

**6. Write a Python program to perform an independent T-test and interpret the results?**

```python
import numpy as np
from scipy import stats

# Step 1: Define two independent sample datasets
group1 = [105, 110, 102, 108, 115, 107, 109, 111, 106, 112]
group2 = [98, 100, 97, 96, 103, 99, 95, 101, 100, 98]

# Step 2: Perform an independent (two-sample) t-test
t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Welch's t-test

# Step 3: Print the results
print("Group 1 Mean:", round(np.mean(group1), 2))
print("Group 2 Mean:", round(np.mean(group2), 2))
print("T-statistic:", round(t_statistic, 3))
print("P-value:", round(p_value, 4))

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("Interpretation: There is a significant difference between the means of the two groups.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference between the means of the two groups.")
```

**7. Perform a paired sample T-test using Python and visualize the comparison results.**

**9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a sample.

    Parameters:
    - data: list or array of numeric sample values
    - confidence: desired confidence level (default is 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval bounds
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    # t critical value (because population std is unknown)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)
sample_data = [105, 110, 102, 108, 115, 107, 109, 111, 106, 112]
ci_95 = confidence_interval(sample_data, confidence=0.95)

print("95% Confidence Interval:", (round(ci_95[0], 2), round(ci_95[1], 2)))
```

```python
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    - data: list or array of numeric sample values
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - margin of error (float)
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    # t-critical value for given confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    moe = t_critical * standard_error
    return moe

# Example usage
sample_data = [105, 110, 102, 108, 115, 107, 109, 111, 106, 112]
moe_95 = margin_of_error(sample_data, confidence=0.95)

print("Margin of Error (95% confidence):", round(moe_95, 2))
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?**

```python
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    """
    Applies Bayes' Theorem to compute P(A|B)

    Parameters:
    - prior_A: P(A) — prior probability of hypothesis A
    - likelihood_B_given_A: P(B|A) — likelihood of observing B if A is true
    - prior_not_A: P(not A) — prior probability of hypothesis not A
    - likelihood_B_given_not_A: P(B|not A) — likelihood of observing B if A is false

    Returns:
    - posterior_A_given_B: P(A|B)
    """
    numerator = prior_A * likelihood_B_given_A
    denominator = (prior_A * likelihood_B_given_A) + (prior_not_A * likelihood_B_given_not_A)
    posterior = numerator / denominator
    return posterior

# Example scenario:
# A medical test for a disease has:
# - 1% disease prevalence (P(Disease)) = 0.01
# - 99% sensitivity (P(Positive | Disease)) = 0.99
# - 95% specificity (P(Negative | No Disease)) → P(Positive | No Disease) = 1 - 0.95 = 0.05

# Inputs
prior_D = 0.01                     # P(Disease)
likelihood_Pos_given_D = 0.99     # P(Positive | Disease)
prior_no_D = 1 - prior_D           # P(No Disease)
likelihood_Pos_given_no_D = 0.05  # P(Positive | No Disease)

# Apply Bayes' Theorem
posterior = bayes_theorem(prior_D, likelihood_Pos_given_D, prior_no_D, likelihood_Pos_given_no_D)

# Output
print("Probability of Disease given a Positive Test (P(Disease | Positive)):", round(posterior, 4))
```

**16. Implement an F-test using Python to compare the variances of two random samples?**

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two independent samples.

    Parameters:
    - sample1, sample2: numeric lists or arrays
    - alpha: significance level (default is 0.05)

    Returns:
    - F-statistic, p-value, and interpretation
    """
    n1, n2 = len(sample1), len(sample2)
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)

    # Always place the larger variance in the numerator
    if var1 > var2:
        F = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1
    else:
        F = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Two-tailed p-value
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

    print("Sample 1 Variance:", round(var1, 3))
    print("Sample 2 Variance:", round(var2, 3))
    print("F-Statistic:", round(F, 4))
    print("P-Value:", round(p_value, 4))

    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis.")
        print("Interpretation: The variances are significantly different.")
    else:
        print("Conclusion: Fail to reject the null hypothesis.")
        print("Interpretation: No significant difference in variances.")

# Example: Simulate two samples
np.random.seed(0)
sample_A = np.random.normal(loc=50, scale=5, size=30)   # variance = 25
sample_B = np.random.normal(loc=55, scale=10, size=30)  # variance = 100

# Run the F-test
f_test(sample_A, sample_B)
```

**19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.**

```python
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    """
    Checks normality and homogeneity of variances for ANOVA.

    Parameters:
    - groups: multiple lists/arrays representing different groups

    Prints:
    - Shapiro-Wilk test results for normality
    - Levene’s test for equal variances
    - Note on independence assumption
    """
    print("🔍 Checking Normality with Shapiro-Wilk Test:")
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        print(f"  Group {i}: p-value = {round(p, 4)} → {'✅ Normal' if p > 0.05 else '❌ Not Normal'}")

    print("\n📏 Checking Homogeneity of Variance with Levene’s Test:")
    levene_stat, levene_p = levene(*groups)
    print(f"  Levene’s Test: p-value = {round(levene_p, 4)} → {'✅ Equal variances' if levene_p > 0.05 else '❌ Unequal variances'}")

    print("\n📎 Independence:")
    print("  ⚠️ Assumed based on study design. Ensure data points are independently sampled.")

# Example usage
np.random.seed(0)
group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=15, scale=2, size=30)
group3 = np.random.normal(loc=20, scale=2, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Generate two datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=30)   # Variance ≈ 25
data2 = np.random.normal(loc=55, scale=10, size=30)  # Variance ≈ 100

# Step 2: Calculate variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Step 3: Compute F-statistic (larger variance in numerator)
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(data1) - 1, len(data2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(data2) - 1, len(data1) - 1

# Step 4: Compute p-value for two-tailed test
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Step 5: Output test results
print("Sample 1 Variance:", round(var1, 2))
print("Sample 2 Variance:", round(var2, 2))
print("F-Statistic:", round(F, 3))
print("P-Value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("Interpretation: The variances are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference in variances.")

# Step 6: Visualization
plt.figure(figsize=(8, 5))
plt.hist(data1, bins=15, alpha=0.6, label='Data 1', color='skyblue')
plt.hist(data2, bins=15, alpha=0.6, label='Data 2', color='salmon')
plt.axvline(np.mean(data1), color='blue', linestyle='dashed', label='Mean 1')
plt.axvline(np.mean(data2), color='red', linestyle='dashed', label='Mean 2')
plt.title('Distribution Comparison of Two Samples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import chisquare

# Step 1: Simulate observed data (e.g., dice rolls)
observed = np.array([18, 22, 16, 20, 24, 20])  # Frequencies of rolling 1-6

# Step 2: Define expected frequencies (fair die → equal probabilities)
expected = np.array([20] * 6)  # Total rolls = 120, expected = 120/6 = 20 for each face

# Step 3: Perform Chi-square goodness of fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print("Chi-Square Statistic:", round(chi2_stat, 3))
print("P-Value:", round(p_value, 4))

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("Interpretation: Observed distribution significantly differs from expected.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference between observed and expected distribution.")
```
